```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sns
import math
filename = "output.csv"
names = ['overlap','x','y','z','i','error']
data = pd.read_csv(filename, names=names)
```

```python
def toArray(var):
    x = []
    for i in range(len(var)):
        x.append(float(var[i]))
    return np.array(x)
```

```python
x = toArray(data.x)
y = toArray(data.y)
z = toArray(data.z)
distance = np.sqrt(x*x+y*y+z*z)
data['distance'] = distance
```

```python
data.boxplot()
plt.show()
```

```python
data.hist()
plt.show()
```

```python
from pandas.tools.plotting import scatter_matrix
scatter_matrix(data, alpha=0.2, figsize=(5, 5), diagonal='kde')
plt.show()
```

```python
d1 = pd.concat([data['error'], data['overlap']], axis=1, keys=['error', 'overlap'])
d1.boxplot(by='overlap')
```

```python
l = []
for index, row in data.iterrows():
    d = row['distance']
    if d < 0.1:
        l.append(0.1)
    elif d < 0.2:
        l.append(0.2)
    elif d < 0.3:
        l.append(0.3)
    elif d < 0.4:
        l.append(0.4)
    elif d < 0.5:
        l.append(0.5)
    elif d < 0.6:
        l.append(0.6)
    elif d < 0.7:
        l.append(0.7)
    elif d < 0.8:
        l.append(0.8)
    elif d < 0.9:
        l.append(0.9)
    elif d > 0.9:
        l.append(1.0)
s = pd.Series(l)
d1 = pd.concat([data['error'], s], axis=1, keys=['error', 'distance'])
d1.boxplot(by='distance')
```
