```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sns
import math
filename = "example.csv"
names = ['overlap','x','y','z','i','error']
data = pd.read_csv(filename, names=names)
```

```python
def toArray(var):
    x = []
    for i in range(len(var)):
        print var[i]
        x.append(float(var[i]))
    return np.array(x)
```

```python
x = toArray(data.x)
y = toArray(data.y)
z = toArray(data.z)
distance = math.sqrt(x*x+y*y+z*z)
data['distance'] = distance

```

```python
data.boxplot(return_type='axes')
plt.show()
```

```python
data.hist()
plt.show()
```

```python
from pandas.tools.plotting import scatter_matrix
scatter_matrix(data, alpha=0.2, figsize=(5, 5), diagonal='kde')
plt.show()
```
